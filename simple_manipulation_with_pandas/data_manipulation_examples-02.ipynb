__Sorting and subsetting__

```python
# sorting

df.sort_values("price").head()
```

```python
# sorting in descending order

df.sort_values("price", ascending=False).head()
```

```python
# sorting by multiple variables

df.sort_values(["price", "beds"]).head()
```

```python
# sorting by multiple variables

df.sort_values(["price", "beds"], ascending=[True,False]).head()
```

```python
# subsetting multiple columns

df[["sale_date","price"]]
```

```python
# subsetting multiple columns using a variable 'col'
# using .head() to visualize only first 5 rows

col = ["sale_date","price"]
df[col].head()
```

```python
# subsetting columns 

df[df['price'] > 90000]
```

```python
# based on multiple conditions 
df['year'] = df['sale_date'].str[-4:]
df['year']
```

```python
# based on multiple conditions 

df['price'] = df['price'].astype(int)
df['date'] = df['year'].astype(int)

price = df['price'] >= 90000
date = df['date'] >= 2008

df[price & date]
```

```python
# Subsetting using .isin()

years = df['year'].isin(['2008','2009'])
df[years]
```

```python
# adding a new column

df["price_100"] = df["price"] / 100
print(df.head())
```

```python
# multiple_manipulation

price_lower_50 = df[df["price_100"] < 600]

df_price_lower_50 = price_lower_50.sort_values("year")
df_price_lower_50[["year","price","type"]].head()
```

```python
# sorting values
# head to limit to firsts 5 rows
df.sort_values("year").head()
```

```python
# sorting values in descending order
# head to limit to firsts 5 rows
df.sort_values("year", ascending=False).head()
```

```python
# sorting by multiple variables
# head to limit to firsts 5 rows
## Here I also filter by Year > 2010
df[df["date"] > 2007].sort_values(["date", "price"])
```

__Summary statistics:__

```python
# summarizing numerical data

df["price"].mean()
```

```python
# summarizing numerical data

df["price"].min()
```

```python
# The .agg() method

def pct30(column):
    return column.quantile(0.3)

df["price"].agg(pct30)
```

```python
# Summaries on multiple columns

df[["price","price_100"]].agg(pct30)
```

```python
# The .agg() method

def pct40(column):
    return column.quantile(0.4)

df["price"].agg([pct30,pct40])
```

```python
# Dropping duplicate names
df.drop_duplicates(subset="year").head()
```

```python
df_unique = df.drop_duplicates(subset=["year","price"])
df_unique
```

```python
df_unique["year"].value_counts(sort=True, ascending=True).head(10)
```

```python
# proportions
df_unique["year"].value_counts(normalize=True)
```

```python
# Grouped summaries
df.groupby("year")["price"].mean()
```

```python
# multiple year summaries
df.groupby("year")["price"].agg([min,max,sum])
```

```python
#Many groups, many summaries

df.groupby(["year","type"])[["price","price_100"]].mean()
```
